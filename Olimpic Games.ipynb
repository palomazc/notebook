```python
import numpy as np
import pandas as pd
import matplotlib
```

```python
file='dictionary.csv'
countries_df=pd.read_csv(file)
countries_df.head()
```

```python
file='summer.csv'
summer_df=pd.read_csv(file)
summer_df.head()
```

```python
summer_df.rename(columns={'Country':'Code'},inplace=True) 
summer_df.head()
```

```python
file='winter.csv'
winter_df=pd.read_csv(file)
winter_df.head()
```

```python
winter_df.rename(columns={'Country':'Code'},inplace=True) 
winter_df.head()
```

```python
summer_countries_df=summer_df.merge(countries_df)
summer_countries_df.head()
```

```python
summer_countries_df['season']='Summer'
summer_countries_df.head()
```

```python
winter_countries_df=winter_df.merge(countries_df)
winter_countries_df.head()
```

```python
winter_countries_df['season']='Winter'
winter_countries_df.head()
```

```python
all_df = pd.concat([summer_countries_df, winter_countries_df], sort=False)
```

```python
all_df.tail()
```

```python
year_1984=all_df['Year']>=1984
all_df_1984=all_df[year_1984]
medalls_df=all_df_1984.groupby('Code').count()[['Medal']]
medalls_ordered_df=medalls_df.sort_values(by='Medal',ascending=False)
medalls_ordered_df.head(10).plot(kind='bar')
```

```python
all_df_medals_season=all_df_1984.groupby(['Code','season'])['Country'].count().unstack()
all_df_medals_season.fillna(0,inplace=True)
all_df_medals_season.head()
all_df_medals_season['Summer']=all_df_medals_season['Summer'].astype(int)
all_df_medals_season['Winter']=all_df_medals_season['Winter'].astype(int)
all_df_medals_season['Total']=all_df_medals_season['Winter']+all_df_medals_season['Summer']
all_df_medals_season.head()
all_df_medals_season=all_df_medals_season.sort_values(by='Total',ascending=False)
all_df_medals_season.head(10).plot(kind='bar')
```
